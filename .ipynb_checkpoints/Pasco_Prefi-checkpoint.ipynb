## Data Preprocess

```python
import os
import arff
import pandas as pd

def preprocess_arff(arff_path):
    """
    Preprocess an ARFF file and return a Pandas DataFrame.

    Parameters:
    - arff_path (str): Path to the ARFF file.

    Returns:
    - pd.DataFrame: Preprocessed DataFrame.
    """
    # Load ARFF file
    with open(arff_path, 'r') as file:
        arff_data = arff.load(file)

    # Convert to a DataFrame
    data = arff_data['data']
    column_names = [attr[0] for attr in arff_data['attributes']]
    df = pd.DataFrame(data, columns=column_names)

    # Convert non-numeric columns (i.e., 'string' or 'nominal') to string type
    for attr in arff_data['attributes']:
        if attr[1] == 'STRING' or isinstance(attr[1], list):
            df[attr[0]] = df[attr[0]].astype(str)

   # Convert 'nan' strings to actual NaN values
    df.replace('m', np.nan, inplace=True)

    return df

# Loop over multiple ARFF files
file_path = './V4 data/'
file_names = [f for f in os.listdir(file_path) if f.endswith('.arff')]
preprocessed_dataframes = []

for file_name in file_names:
    arff_path = os.path.join(file_path, file_name)
    preprocessed_df = preprocess_arff(arff_path)
    preprocessed_dataframes.append(preprocessed_df)

# Concatenate the preprocessed DataFrames
combined_df = pd.concat(preprocessed_dataframes, ignore_index=True)

print(combined_df)
```

```python
# Remove NaN values and remove Num and Country column
combined_df2 = combined_df.dropna()
combined_df2 = combined_df2.drop(['Num', 'Country'], axis=1)

print(combined_df2)
```

```python
#Convert DataFrame to matrix
A = combined_df2.values
A = A.astype(float)
print(A)
```

# PCA

```python
from sklearn import preprocessing

class myPCA:

    def __init__(self, n_components):
        self.n_components = n_components
        self.components = None
        self.mean = None

    def fit(self, X):
        # mean centering
        self.mean = np.mean(X, axis=0)
        X = X -  self.mean

        # covariance, functions needs samples as columns
        cov = np.cov(X.T)

        # eigenvectors, eigenvalues
        eigenvectors, eigenvalues = np.linalg.eig(cov)

        # eigenvectors v = [:, i] column vector, transpose this for easier calculations
        eigenvectors = eigenvectors.T

        # sort eigenvectors
        idxs = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idxs]
        eigenvectors = eigenvectors[idxs]

        self.components = eigenvectors[:self.n_components]

    def transform(self, X):
        # projects data
        X = X - self.mean
        return np.dot(X, self.components.T)
```

```python
PCA_A = preprocessing.scale(A);
mypca = myPCA(n_components=83)
mypca.fit(PCA_A)
myA_projected = mypca.transform(PCA_A)

print("Shape of A:", A.shape)
print("Shape of transformed A:", myA_projected.shape)
```
